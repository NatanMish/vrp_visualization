```python
def plot_path(points, path_color):
    fig, ax = plt.subplots(figsize=(10,10))
    
    codes = [
    Path.MOVETO,
    ]
    
    [codes.append(Path.LINETO) for _ in range(len(points[1:]))]

    path = Path(points, codes)

    # plot control points and connecting lines
    x, y = zip(*path.vertices)
    line, = ax.plot(x, y, 'o-', lw=10, color=path_color, ms=30)
    ax.plot(points[-1][0], points[-1][1], 'o', lw=10, color='red', ms=30)

    for i, point in enumerate(points[:-1]):
        ax.text(point[0], point[1], i, color='green', size=75)
        
    ax.text(points[-1][0], points[-1][1], 'Destination', color='red', size=75)

    ax.grid()
    ax.axis('equal')
    
    distance_covered = round(sum(math.dist(points[i], points[i+1]) for i in range(len(points)-1)), 2)
    fig.suptitle(f'Distance covered: {distance_covered}', fontsize=26)
    plt.savefig('points_4_1.png')
    plt.show()
```

```python
def create_permutations(points):
    perm = permutations(points[:-1])
    combinations = []
    for per in perm:
        path = list(per)
        path.append(points[-1])
        combinations.append(path)
    
    return combinations
```

```python
def greedy_algorithm_steps(points):
    distance_mat = np.array(DistanceMetric.get_metric('manhattan').pairwise(points))
    furthest_point_from_dest_position = np.argmax(distance_mat[-1])
    np.fill_diagonal(distance_mat, 99999999)
    steps = [[furthest_point_from_dest_position]]
    for step in range(len(points) - 2):
        last_point_position = steps[-1][-1]
        next_point_candidates = np.argsort(distance_mat[last_point_position])
        
        for i in next_point_candidates:
            if i not in steps[-1] and i != (len(points) - 1):
                next_point_poisiton = i
                break
                
        next_step = steps[-1].copy()
        next_step.append(next_point_poisiton)
        steps.append(next_step)
        
    final_step = steps[-1]
    final_step.append(len(points) - 1)
    steps.append(final_step)
    
    steps_coordinates = []
    
    for step in steps:
        step_coordinates = []
        for i in step:
            step_coordinates.append(points[i])
        steps_coordinates.append(step_coordinates)
        
    return steps_coordinates
```

```python
def two_opt_algorithm_steps(points):
    distance_mat = np.array(DistanceMetric.get_metric('manhattan').pairwise(points))
    def cost_change(cost_mat, n1, n2, n3, n4):
        return cost_mat[n1][n3] + cost_mat[n2][n4] - cost_mat[n1][n2] - cost_mat[n3][n4]


    def two_opt(route, cost_mat):
        best = route
        improved = True
        steps = []
        while improved:
            improved = False
            for i in range(1, len(route) - 2):
                for j in range(i + 1, len(route)):
                    if j - i == 1: continue
                        
                    best_copy = best.copy()
                    best_copy[i:j] = best[j - 1:i - 1:-1]
                    steps.append(best_copy)
                    if cost_change(cost_mat, best[i - 1], best[i], best[j - 1], best[j]) < 0:
                        best[i:j] = best[j - 1:i - 1:-1]
                        improved = True
            route = best
        return best, steps
    
    best_route, steps = two_opt(list(range(len(points))), distance_mat)
    
    steps_coordinates = []
    
    for step in steps:
        step_coordinates = []
        for i in step:
            step_coordinates.append(points[i])
        steps_coordinates.append(step_coordinates)
        
    return steps_coordinates
```
